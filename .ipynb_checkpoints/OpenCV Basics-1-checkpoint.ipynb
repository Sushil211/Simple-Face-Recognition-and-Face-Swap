```python
import cv2
import matplotlib.pyplot as plt
```

```python
#Read an image
img = cv2.imread("dog.png")
```

```python
plt.imshow(img)
plt.show()
```

<p><b>Look at the original image and this image, the color looks somewhat different. The reason behind this is that, in real the image is stored in 3 channels (R,G,B), but by-default openCV reads the image as 3 color channels (B,G,R), so that means the R(Red) and B(Blue) color channels are getting swapped when we try to read an image using openCV. So we need to switch these colors back to their original position.</b></p>

```python
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
#So this is basically a constant in the library which is mapped to certain integer.
#So when OpenCv sees this constant, it basically understands that which type of conversion we want to do.
```

```python
plt.imshow(newImg)
plt.show()
```
